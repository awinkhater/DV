## Question ($\star\star$, 30 marks)
- Problem: 
    - follow up on Exercise 3.15 in Chapter 3 of lecture note
    - find a target in a $m \times n$ matrix
    - here we can assume that 
        - each row in the matrix is sorted in ascending order
        - each column in the matrix is sorted in ascending order
        - $m < n$
    - see examples in the testing cases
    - find the solution with the complexity below
- Complexity: 
    - $O(m \times log(n)$ time
    - $O(1)$ space
- **Logic: # Implement me (5 marks)**


```python
# Implementation
def fun_21(matrix, target):
    """
    Find a target in a m * n matrix
    
    Parameters
    ----------
    matrix : list[list[int]] (i.e., a list of list of integers)
    target : an integer
    
    Returns
    ----------
    True  : if the target is in the matrix
    False : otherwise
    """
    
    # Implement me (25 marks)
    #Logit
    #THIS FUNCTION WORKS BY Sorting through the rows of the matrix

    # Corner case
    # If matrix is any type of empty, this function will return False
    if len(matrix) == 0 or len(matrix[0]) == 0:
        return False

    # Find values for M and n
    m, n = len(matrix), len(matrix[0])
    
    for i in range(m):
        # Binary search over the columns of the matrix
        left, right = 0, n - 1
        while left <= right:
            # Get the middle column
            mid = left + (right - left) // 2
            #If the Target is found, return True
            if matrix[i][mid] == target:
                return True
            #Keep Looking using the ascendinng order by reducing the right most bound
            elif matrix[i][mid] > target:
                right = mid - 1
            #Same thing but for the left case
            else:
                left = mid + 1
            
    return False
#The time complexity for this function is O(m* log(n)) because it has a for loop dependent on m and it is using a binary search over the n paramaeter
#The space complexity for this function O(1) because no values are being copied and the amount of space taken is not dependent on the input matrix
```

## Question ($\star\star$, 20 marks)
- Problem: 
    - follow up on Exercise 5.5 in Chapter 5 of lecture note
    - find whether there is a path in the binary tree such that the sum of value of nodes in the path equals a target
    - here, the path starts from the root and ends at a leaf node
    - see examples in the testing cases
    - find the solution with the complexity below    
    - you must use DFS to design the solution
- Complexity: 
    - $O(n)$ time
    - $O(n)$ space
    - where $n$ is the number of nodes in the tree
- **Logic: # Implement me (5 marks)**


```python
# Implementation
def fun_31(root, target):
    """
    Find whether there is a path in the binary tree such that the sum of value of nodes in the path equals a target
    
    Parameters
    ----------
    root : the root of a binary tree
    target : an integer
    
    Returns
    ----------
    True,  if there is such a path
    False, otherwise
    """
    
    # Implement me (15 marks)
    #Logit
    #We use a similar function to fun_55 to log the value of the path functions. If there are applicable path sums
    #They will get added to this global paths list.
    global paths      
    paths = []
#Here we call our helper function
    helper(root, 0, target)
    #Here we check if there are any satisfactory sums found
    if len(paths) > 0:
        return (True)
    else:
        return False 
    

def helper(root, path, target):
    """
    Find the paths (from the root to a leaf) of a binary tree   
    and add them to the paths list if they are equal to the target
    Parameters
    ----------
    root : the root of a binary tree
    
    Returns
    ----------
    The paths that satisfy the target
    """

    # Base case
    if root == None:
        return

    # Base
    if root.left == None and root.right == None:
        # Add the value of the root to path        
        path += root.val
        # Append path to paths if it is satisfactory
        if path == target:
            paths.append(path)
        else:
            return paths
        return
    # Recursion
    else:
        path += root.val   
        helper(root.left, path, target)
        helper(root.right, path, target)

#The time complexity of this function is O(n) as we only traverse the tree from root to leaves once per path
#The space complexity of this function is O(n) because we create a copy of the paths list
```

## Question ($\star$, 15 marks)
- Problem: 
    - follow up on Exercise 5.10 in Chapter 5 of lecture note
    - find the number of (vertically or horizontally) connected deserts in a map
    - unlike Exercise 5.10 where the map after running the solution does not have to be the same as the input one, here the map must be the same as the input
    - that is, you must reverse any change made in the map after DFS <font color='red'>(you may consider this as a hint)</font>
    - see examples in the testing cases
    - find the solution with the complexity below   
    - note that you cannot make a copy of the map since that requires $O(n)$ space (where $n$ is the number of entries in the map)
- Complexity: 
    - $O(n^2)$ time
    - $O(1)$ space
- **Logic: # Implement me (5 marks)**


```python
# Reference: Part of the code below is from chapter_5_tree_search_solution.ipynb
# Implementation
def fun_32(matrix):
    """
    Find the number of (vertically or horizontally) connected deserts in a map (represented by a matrix)
    
    Parameters
    ----------
    matrix : a matrix where 1 denotes desert and 0 denotes grassland
    
    Returns
    ----------
    the number of connected deserts
    """

    # Corner case
    if len(matrix) == 0 or len(matrix[0]) == 0:
        return 0

    m, n = len(matrix), len(matrix[0])
    # The number of connected deserts
    count = 0

    for i in range(m):
        for j in range(n):
            if matrix[i][j] == 1:
                count += 1
                # DFS: remove '1's conntected to entry [i, j]
                helper2(matrix, m, n, i, j)
                
    # Implement me (5 marks)
    # Restore the matrix
    #Logit: This function works very similarly to the original function in 5-10, but with two key differences:
    #1. This function replaces 1's with -1 to differentiate them from grasslands
    #2 This function loops over itself again after the count is found to restore all the -1's to 1s

    #The negative 1's allow us to use placeholder values that are not confused with grasslands' 0

    #This is simply the reverse the changes loop
    for i in range(m):
        for j in range(n):
            if matrix[i][j] == -1:
                matrix[i][j]=1
                
                
    
    return count
```

```python
def helper2(matrix, m, n, i, j):
    """
    DFS: remove '1's conntected to entry [i, j]

    Parameters
    ----------
    matrix : a matrix where 1 denotes desert and 0 denotes grass land
    m : the nubmer of rows in the matrix
    n : the nubmer of columns in the matrix
    i, j : entry [i, j]
    """

    # Implement me (5 marks)
    if i < 0 or i >= m or j < 0 or j >= n or matrix[i][j] != 1:
        return

    # Remove 1's and place them with -1's

    if matrix[i][j] == 1:
        matrix[i][j]= -1 

#The 4 directional calls remain unaltered
    # Remove '1' above [i, j]
    helper2(matrix, m, n, i - 1, j)
    # Remove '1' below [i, j]
    helper2(matrix, m, n, i + 1, j)
    # Remove '1' on the left-hand side of [i, j]
    helper2(matrix, m, n, i, j - 1)
    # Remove '1' on the right-hand side of [i, j]
    helper2(matrix, m, n, i, j + 1)


#TIME COMPLEXITY: The time complexity of this function is O(n) as it loops over the input twice: Once to count deserts and once to reverse changes made
#SPACE COMPLEXITY: The space complexity of this function is O(1) as the only thing that is stored is the count variable.
```

# Breadth First Search (35 marks)

## Question ($\star\star$, 20 marks)
- Problem: 
    - follow up on Question 3.1 above
    - find whether there is a path in the binary tree such that the sum of value of nodes in the path equals a target
    - see examples in the testing cases
    - find the solution with the complexity below
    - you must use BFS to design the solution
    - the node value can be modified <font color='red'>(you may consider this as a hint)</font>
- Complexity: 
    - $O(n)$ time
    - $O(n)$ space
- **Logic: # Implement me (5 marks)**


```python
# Implementation
from collections import deque

def fun_41(root, target):
    """
    Find whether there is a path in the binary tree such that the sum of value of nodes in the path equals a target
    
    Parameters
    ----------
    root : the root of a binary tree
    target : an integer
    
    Returns
    ----------
    True,  if there is such a path
    False, otherwise
    """

    # Implement me (15 marks)
    #Logit:
    #The way I am approaching this problem is to save the value of the current node in a 'vals' variable and then
    #add it to the value for both its left and right nodes and so forth so that the leaves will have the values of their entire path
    #We are using basic BFS queue to process the tree

    #Here is the initial value for the value
    vals = 0
    
    # Corner case
    if root == None:
        return False
    
    # Initialize the queue
    queue = deque([root])
    # BFS
    while len(queue) > 0:
        # Get the length of the queue
        n = len(queue)
        for _ in range(n):
            # Remove the first node in the queue
            node = queue.popleft()
            #Set the value counter to the value of the first node
            vals=node.val
            #Leaf condition that checks
            if node.left == None and node.right == None:
                if vals == target:
                    return True
            else:
                if node.left != None:
                    # Add the left and right children to the end of the queue and change the node value 
                    node.left.val+= vals
                    queue.append(node.left)
                if node.right != None:
                    node.right.val+= vals
                    # Add the left and right children to the end of the queue change the node value 
                    queue.append(node.right)
            #Return False if true condition wasnt hit on the leaves  
    return False

#The time complexity of this function is O(n) as it iterates through the tree only once
#The space complexity of this function is O(n) as it stores the queue as a copy of the input
```

## Question ($\star$) (15 marks)
- Problem:
    - follow up on Exercise 6.10 in Chapter 5 of lecture note
    - unlike Exercise 6.10 where we should find the bottom-left node in a binary tree, here we should find the bottom-right node
    - see examples in the test cases
    - find the solution with the complexity below
    - you must use BFS to design the solution
- Complexity:
    - $O(n)$ time
    - $O(n)$ space
- **Logic: # Implement me (5 marks)**


```python
# Implementation
def fun_42(root):
    """
    Find the bottom-right node in a binary tree
    
    Parameters
    ----------
    root : the root of a tree
    
    Returns
    ----------
    The value of the bottom-right node in a binary tree
    """

    # Implement me (10 marks)
    # Logit
    #The method we take to solve this function is to iterate through a level of a tree to the right most element
    #Once we find the rightmost element, we proceed down to the bottom row to find the most right object in that row
    
    #Corner Case
    if root == None:
        return root
    #set Bottom Right Variable
    bottom_right = root.val
    
    # Initialize the queue
    queue = deque([root])
    
    while len(queue) > 0:
        # Get the length of the queue
        n = len(queue)
        for i in range(n):
            # Remove the first node in the queue
            node = queue.popleft()

            # Update bottom_right starting with the left most value
            if i == (0):
                bottom_right = node.val

            if node.right != None:
                # Add the value of the right child to the end of the queue
                queue.append(node.right)
                #If there is no right child, we move onto the left child of the rightmost node
            if node.left != None:
                # Add the value of the left child to the end of the queue
                queue.append(node.left)

    return bottom_right
#Time Complexity of this function is O(n) because we iterate through the tree only once
#Space Complexity of this function is O(n) because we create a copy of the tree with the queue
```
